# 辞書型、Json型
1. 上表で、埼玉県川越市から埼玉県秩父市までの人口を、key:市名　value:人口　として辞書を作成せよ
2. この辞書に埼玉県所沢市の人口を追加せよ
3. 埼玉県川越市の人口を取り出せ
4. 埼玉県熊谷市の人口を190000に変更せよ
5. for文を使って、この辞書を先頭から取り出して、key と valueを表示せよ
 


```python
saitama_dic = {'埼玉県川越市':350745,'埼玉県熊谷市':198742,'埼玉県川口市':578112,'埼玉県行田市':82113,'埼玉県秩父市':24038}
print(saitama_dic)
saitama_dic['埼玉県所沢市']=340386
val = saitama_dic['埼玉県川越市']
print(val)
saitama_dic['埼玉県熊谷市']=190000
for k,v in saitama_dic.items():
    print(k,v)
```

6. 上記の辞書のすべてのvalueに人口だけでなくて、世帯数も追加せよ。（人口、制対数を要素とするリスト型にすればよい）
7. この辞書の要素をfor文で先頭から取り出して表示せよ
8. 埼玉県行田市の世帯数を表示せよ

```python
saitama_dic['埼玉県川越市']=[350745,145715]
saitama_dic['埼玉県熊谷市']=[198742,77004]
saitama_dic['埼玉県川口市']=[578112,245830]
saitama_dic['埼玉県行田市']=[82113,31015]
saitama_dic['埼玉県秩父市']=[63555,24038]
saitama_dic['埼玉県所沢市']=[340386,145834]
for k,v in saitama_dic.items():
    print(k,v)
print(saitama_dic['埼玉県行田市'][1])
```

8. 上表では、埼玉県さいたま市のみ、区毎に人口等が表示されている。そこで、上記の辞書型に、  
        keyさいたま市 : value{大宮区:[人口,世帯数,...]  
    のように、辞書を階層型にしてさいたま市の人口、世帯数を反映させよ。なお、対象は、西区から大宮区とする。
9. 作成した辞書から、さいたま市大宮区の人口、世帯数を表示せよ
                                                              

```python
subdic = {'西区':[87146.33871],'北区':[143446,61756],'大宮区':[113864,53265]}
saitama_dic['埼玉県さいたま市']=subdic
print(saitama_dic)
print(saitama_dic['埼玉県さいたま市']['大宮区'])
```
